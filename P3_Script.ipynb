## Creating a map for Antarctic Continent and Southern Ocean from 60Â°S


```python
import cartopy.crs as ccrs
import cartopy
import matplotlib.pyplot as plt
import cartopy.feature as cfeature
import numpy as np
```

```python
extent = [-180, 180, -90, -60]

plt.figure(figsize=(10, 5))
ax = plt.axes(projection=ccrs.SouthPolarStereo())
ax.set_title('Antarctica & the Southern Ocean')
ax.set_extent(extent, ccrs.PlateCarree())

ax.add_feature(cfeature.OCEAN, color = "lightblue")
ax.add_feature(cartopy.feature.LAND, color = "white", edgecolor="black")
ax.coastlines("50m", color = "black")
ax.gridlines(draw_labels=True)
```

Figure 1. Map indicating the circumpolar extent of the Southern Ocean and the Antarctic Continent between 60 and 90 degrees S. The South Polar Stereo coordinate reference system was utilised as this projection allowed me to have Antarctica centralised on the map and indicate the circumpolar feature of the Southern Ocean.

## Creating a map of the South Atlantic, from 20Â°S to 50Â°S

```python
# Coordinates for locations Walvis Bay, Cape Town, Rio de Janeiro, Montevideo
walvis = dict(lon=14.50528, lat=-22.9575)
cpt = dict(lon=18.42322, lat=-33.92584)
rio = dict(lon=-43.18223, lat=-22.90642)
montevideo = dict(lon=-56.164532,lat= -34.901112)
lons = [walvis['lon'], cpt['lon'], rio['lon'], montevideo['lon']]
lats = [walvis['lat'], cpt['lat'], rio['lat'], montevideo['lat']]
```

```python
extent = [-80, 20, -20, -60]

#place = ['Walvis', 'Cape Town', 'Rio de Janeiro', 'Montevideo']
plt.figure(figsize=(12, 6))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.set_title('The South Atlantic Ocean & locations of four cities')
ax.set_extent(extent)
ax.scatter(lons[0], lats[0], label='Walvis', color = 'red', transform = ccrs.PlateCarree())
ax.scatter(lons[1], lats[1], label='Cape Town', color = 'blue', transform = ccrs.PlateCarree())
ax.scatter(lons[2], lats[2], label='Rio de Janeiro', color = 'green', transform = ccrs.PlateCarree())
ax.scatter(lons[3], lats[3], label='Montevideo', color = 'purple', transform = ccrs.PlateCarree())
ax.legend()
ax.add_feature(cfeature.OCEAN)
ax.add_feature(cartopy.feature.LAND, edgecolor='black')
ax.add_feature(cartopy.feature.LAKES, edgecolor='black')
ax.add_feature(cartopy.feature.RIVERS)
ax.gridlines(draw_labels=True)
```

Figure 2. Map indicating the South Atlantic Ocean between 80 degrees W, 20 degrees E and 20 degrees S, 70 degrees S. Using the Plate Carree projection did not allow for the extent to be exactly between 20 and 60 degrees S. This reference system was utilised as a relatively large area was mapped and smaller area features did not need to be accurate. The four city locations are indicated by the legend.

## Three panels indicating the False Bay region in different resolutions

```python
cl_low = cfeature.GSHHSFeature(scale='l')
cl_inter = cfeature.GSHHSFeature(scale='i')
cl_high = cfeature.GSHHSFeature(scale='h')

extent = [18, 19, -34, -34.5]

fig,ax = plt.subplots(1,3,figsize=(10,6),subplot_kw={'projection':ccrs.Orthographic()})
ax[0].set_extent(extent)#, ccrs.PlateCarree())
gl0 = ax[0].gridlines(draw_labels=True)
ax[0].add_feature(cl_low)

ax[1].set_extent(extent)#,ccrs.PlateCarree())
gl1 = ax[1].gridlines(draw_labels=True)
ax[1].add_feature(cl_inter)

ax[2].set_extent(extent)#,ccrs.PlateCarree())
gl2 = ax[2].gridlines(draw_labels=True)
ax[2].add_feature(cl_high)

gl0.right_labels = False
gl0.top_labels = False
gl1.right_labels = False
gl1.top_labels = False
gl2.right_labels = False
gl2.top_labels = False

# set the spacing between subplots and title
plt.suptitle('False Bay maps with different resolutions', x=0.5, y=0.65)
fig.tight_layout()
plt.show()
```

Figure 3. Maps showing the False Bay region using three different resolutions, low, intermediate and high (from left to right respectively). The orthographic projection was utilsed for this map as it is very zoomed in and a more accurate Coordinate Reference System is needed.
